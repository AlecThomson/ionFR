```python
#Simple plot to view ionFR output

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# read text file output IonRM.txt
hour, tec, bfield, rm_ion, rm_ion_err = np.loadtxt('../ionfr/data/tests/IonRM.txt',unpack=True)
print(len(hour))
ion_df = pd.read_csv("../ionfr/data/tests/IonRM.csv")
ion_df
```

```python
# plot numbers from text file
fig = plt.figure(figsize=(8,6), dpi=250)
params = {'text.usetex': True, 'axes.labelsize': 18, 'font.size': 18}
plt.rcParams.update(params)

plt.errorbar(hour,rm_ion,yerr=rm_ion_err,fmt='ro',ms=5, label="Old IonFR")
plt.errorbar(ion_df["hour"],ion_df["ifr"],yerr=ion_df["rms_ifr"],fmt='bo',ms=5, label="New IonFR")
plt.xlabel('$\mathrm{Time~(h~UT)}$',fontsize=20)
plt.ylabel('$\mathrm{RM_{ion}~(rad~m^{-2})}$',fontsize=20)
plt.legend()
```
